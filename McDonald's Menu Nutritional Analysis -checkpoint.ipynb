```python
import pandas as pd

# Load the dataset
data = pd.read_csv('menu.csv')  # replace with actual file path
data
```

```python
# Inspect the first few rows
print(data.head())
```

```python
# Check for basic info
print(data.info())
```

```python
# Check for missing values
print(data.isnull().sum())

# Handle missing values (e.g., drop or fill)
data = data.dropna()  # Alternatively, use data.fillna(value) for filling

# Check for duplicate entries
data = data.drop_duplicates()

# Inspect data types and convert if necessary
data['Calories'] = pd.to_numeric(data['Calories'], errors='coerce')
```

```python
# Summary statistics
calorie_stats = data['Calories'].describe()
print(calorie_stats)
```

```python
# Nutritional content overview
nutritional_columns = ['Total Fat', 'Protein', 'Carbohydrates']
nutritional_stats = data[nutritional_columns].describe()
print(nutritional_stats)

# Check correlation between nutritional components
correlation = data[nutritional_columns].corr()
print(correlation)
```

```python
# Group by food categories
category_trends = data.groupby('Category')[nutritional_columns].mean()
print(category_trends)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histogram of calorie counts
plt.figure(figsize=(10, 6))
sns.histplot(data['Calories'], bins=30, kde=True)
plt.title('Distribution of Calorie Counts')
plt.xlabel('Calories')
plt.ylabel('Frequency')
plt.show()

# Box plot for nutritional content by category
plt.figure(figsize=(12, 8))
sns.boxplot(data=data, x='Category', y='Calories')
plt.title('Calorie Counts by Menu Category')
plt.xticks(rotation=45)
plt.show()
```

```python
# Bar plots for average nutritional content by category
avg_nutrition = data.groupby('Category')[nutritional_columns].mean().reset_index()

plt.figure(figsize=(12, 8))
avg_nutrition.plot(kind='bar', x='Category')
plt.title('Average Nutritional Content by Menu Category')
plt.ylabel('Nutritional Values')
plt.show()
```

```python
highest_calories = data.loc[data['Calories'].idxmax()]
lowest_calories = data.loc[data['Calories'].idxmin()]

print(f"Highest Calorie Item:\n{highest_calories}")
print(f"Lowest Calorie Item:\n{lowest_calories}")
```

```python
popular_categories = data['Category'].value_counts().nlargest(5).index
average_nutrition_popular = data[data['Category'].isin(popular_categories)].groupby('Category')[nutritional_columns].mean()
print(average_nutrition_popular)
```
